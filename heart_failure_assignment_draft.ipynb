#### Removing rows or columns with too many missing values

```python
# Set a threshold for the maximum percentage of missing values allowed
column_threshold = 0.5  # 50%
row_threshold = 0.5  # 50%

# Drop columns with more than 50% missing values
df_few_nans = df.dropna(thresh=int((1 - column_threshold) * len(df)), axis=1)

# Drop rows with more than 50% missing values
df_few_nans = df_few_nans.dropna(thresh=int((1 - row_threshold) * len(df_few_nans.columns)), axis=0)

# Display the shape of the cleaned dataframe
df_few_nans.shape
```

```python
# Checking remaining columns with their Nans
nan_counts = df_few_nans.isna().sum()
nan_counts_sorted = nan_counts.sort_values(ascending=False)
nan_counts_sorted
```

```python
df_few_nans.head(10)
```

#### Filling missing values with the mean

```python
# # Filling missing values with the mean of each column
# df_filled = df_few_nans.fillna(df.mean())
# # Display the first few rows to confirm the filling
# df_filled.head(10)
```

```python
import pandas as pd

# Assuming df_few_nans is your DataFrame

# Step 1: Calculate the mean only for numeric columns
numeric_means = df_few_nans.select_dtypes(include=['number']).mean()

# Step 2: Fill missing values in numeric columns with their respective means
df_filled = df_few_nans.fillna(numeric_means)

# Step 3: Display the first few rows to confirm the filling
print(df_filled.head(10))
```

#### Type of remaining variables

```python
# Display all columns and their data types
df_filled.info()
```

We still have 2 categorical variables with Nans : Primary Insurance & Race Simplified

```python
# Impute missing values in 'Primary Insurance' with the mode (most common value)
primary_insurance_mode = df_filled['Primary Insurance'].mode()[0]
df_filled['Primary Insurance'].fillna(primary_insurance_mode, inplace=True)
```

```python
# For 'Race Simplified', you can either fill with the mode or set to 'Unknown'
race_simplified_mode = df_filled['Race Simplified'].mode()[0]
df_filled['Race Simplified'].fillna('Unknown', inplace=True)
```

#### Model 1: Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score


# Train a RandomForest Classifier
rf_model = RandomForestClassifier(random_state=42)

# Fit the model on the entire training data
rf_model.fit(X_train, y_train)
```

### Model Evaluation

- ROC (Receiver Operating Characteristic) Curve
-  AUC (Area Under the Curve)

```python
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, confusion_matrix, accuracy_score
# Predictions
y_pred = rf_model.predict(X_test)
y_prob = rf_model.predict_proba(X_test)[:, 1]

# Evaluation metrics
print(classification_report(y_test, y_pred))
```

```python
# Calculate and print Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
```

```python
import matplotlib.pyplot as plt

# Step 7: Calculate ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

# Plotting the ROC curve
plt.figure(figsize=(8, 6))  # Optional: Adjusts the size of the plot
plt.plot(fpr, tpr, label=f'RF AUC = {auc_score:.3f}', marker='.')
plt.plot([0, 1], [0, 1], 'k--')  # Adds a diagonal dashed line representing random chance

# Adding labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - RandomForest')
plt.legend(loc='lower right')  # Adds a legend with the AUC label positioned in the lower right

# Display the plot
plt.show()
```

### Model 5: Neural Network (Deep Learning)

```python
# pip install tensorflow
```

```python
#import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Initialize the neural network model
nn_model = Sequential()

# Step 2: Add layers with improvements
nn_model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))  # Input layer with more neurons
nn_model.add(BatchNormalization())  # Batch normalization
nn_model.add(Dropout(0.3))  # Dropout to prevent overfitting

nn_model.add(Dense(64, activation='relu'))  # First hidden layer with more neurons
nn_model.add(BatchNormalization())  # Batch normalization
nn_model.add(Dropout(0.3))  # Dropout to prevent overfitting

nn_model.add(Dense(32, activation='relu'))  # Second hidden layer
nn_model.add(BatchNormalization())  # Batch normalization
nn_model.add(Dropout(0.3))  # Dropout to prevent overfitting

nn_model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Step 3: Compile the model with a smaller learning rate
nn_model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0001), metrics=['accuracy'])

# Step 4: Create a checkpoint callback to save the best model and track the best accuracy
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]


# Step 5: Train the model with the checkpoint callback
history = nn_model.fit(X_train, y_train, epochs=2000, batch_size=32, validation_split=0.2, verbose=1, callbacks=callbacks_list)
```

```python
# Step 6: Predict outcomes
y_prob_nn = nn_model.predict(X_test).ravel()  # Predicted probabilities
y_pred_nn = (y_prob_nn > 0.5).astype(int)  # Convert probabilities to binary predictions

# Step 7: Calculate accuracy
accuracy_nn = accuracy_score(y_test, y_pred_nn)
print("Neural Network Accuracy:", accuracy_nn)

# Step 8: Draw Confusion Matrix
conf_matrix_nn = confusion_matrix(y_test, y_pred_nn)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nn, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Neural Network')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Step 9: Print the best validation accuracy during training
best_accuracy = max(history.history['val_accuracy'])
best_epoch = history.history['val_accuracy'].index(best_accuracy) + 1  # Epochs start at 1
print(f"Best Validation Accuracy: {best_accuracy} at Epoch: {best_epoch}")
```
